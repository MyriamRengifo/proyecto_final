```python
# UNIFICAMOS JSON EN UN SOLO CSV

import os

# Ruta de la carpeta donde están los archivos JSON
carpeta = "/content/drive/MyDrive/Colab Notebooks/review-Florida"

# Crear una lista para almacenar los DataFrames
dataframes = []

# Iterar sobre los archivos JSON en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith(".json"):  # Asegurarse de que sea un archivo JSON
        ruta_completa = os.path.join(carpeta, archivo)
        try:
            # Leer el archivo JSON como DataFrame (formato línea por línea)
            df = pd.read_json(ruta_completa, lines=True)
            dataframes.append(df)
        except ValueError as e:
            print(f"Error al leer el archivo {archivo}: {e}")

# Concatenar todos los DataFrames
if dataframes:  # Verificar si hay DataFrames en la lista
    df_unificado = pd.concat(dataframes, ignore_index=True)

    # Exportar a un archivo CSV con escapechar
    ruta_csv = os.path.join(carpeta, "review_estado_florida.csv")
    df_unificado.to_csv(ruta_csv, index=False, escapechar="\\")  # Agregar un carácter de escape
    print(f"Archivos unificados y exportados a {ruta_csv}")
else:
    print("No se pudieron leer archivos JSON correctamente.")
```
