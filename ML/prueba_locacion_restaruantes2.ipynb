```python
restaurant_lat_long_puntuacion = restaurant[['id_nombre','latitud','longitud','puntuacion_usuarios']]
lugares_interes_lat_long = lugares_interes[['id_nombre','latitud','longitud']]
```

```python
# Hagamos una comparacion entre los puntos de interes turistico, los restaurantes 'buenos' y los 'malos'.

import pandas as pd
import folium

# Cargar los datasets
# Supongamos que tus datasets se llaman 'restaurantes.csv' y 'puntos_interes.csv'
# df_restaurantes = pd.read_csv('restaurantes.csv')  # Columnas: id, latitud, longitud, puntuacion
# df_puntos_interes = pd.read_csv('puntos_interes.csv')  # Columnas: id, latitud, longitud

# Clasificar restaurantes en buenos y no buenos
restaurant_lat_long_puntuacion['tipo'] = restaurant_lat_long_puntuacion['puntuacion_usuarios'].apply(lambda x: 'Bueno' if x > 4.5 else 'No Bueno')

# Crear el mapa base
# Calculamos un punto promedio para centrar el mapa
centro_mapa = [restaurant_lat_long_puntuacion['latitud'].mean(), restaurant_lat_long_puntuacion['longitud'].mean()]
mapa = folium.Map(location=centro_mapa, zoom_start=12)

# Agregar los puntos de interés al mapa
for _, row in lugares_interes_lat_long.iterrows():
    folium.Marker(
        location=[row['latitud'], row['longitud']],
        popup=f"Punto de Interés ID: {row['id_nombre']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(mapa)

# Agregar los restaurantes buenos al mapa
for _, row in restaurant_lat_long_puntuacion[restaurant_lat_long_puntuacion['tipo'] == 'Bueno'].iterrows():
    folium.Marker(
        location=[row['latitud'], row['longitud']],
        popup=f"Restaurante Bueno ID: {row['id_nombre']} - Puntuación: {row['puntuacion_usuarios']}",
        icon=folium.Icon(color='green', icon='cutlery')
    ).add_to(mapa)

# Agregar los restaurantes no buenos al mapa
for _, row in restaurant_lat_long_puntuacion[restaurant_lat_long_puntuacion['tipo'] == 'No Bueno'].iterrows():
    folium.Marker(
        location=[row['latitud'], row['longitud']],
        popup=f"Restaurante No Bueno ID: {row['id_nombre']} - Puntuación: {row['puntuacion_usuarios']}",
        icon=folium.Icon(color='red', icon='cutlery')
    ).add_to(mapa)

# Guardar el mapa en un archivo HTML
mapa.save('mapa_turistico.html')
print("Mapa guardado como 'mapa_turistico.html'. Ábrelo en tu navegador.")
```

```python
# Vamos a ver si existe relacion entre los lugares de interes turistico y si un restaruant es malo o bueno.

import pandas as pd
import numpy as np
from geopy.distance import geodesic

# Cargar los datasets
# df_restaurantes = pd.read_csv('restaurantes.csv')  # Columnas: id, latitud, longitud, puntuacion
# df_puntos_interes = pd.read_csv('puntos_interes.csv')  # Columnas: id, latitud, longitud

# Clasificar restaurantes en buenos y no buenos
# df_restaurantes['tipo'] = df_restaurantes['puntuacion'].apply(lambda x: 'Bueno' if x > 4.5 else 'No Bueno')

# Función para verificar si un restaurante está cerca de algún punto de interés
def esta_cerca(lat_restaurante, lon_restaurante, puntos_interes, distancia_maxima):
    for _, punto in puntos_interes.iterrows():
        distancia = geodesic((lat_restaurante, lon_restaurante), (punto['latitud'], punto['longitud'])).meters
        if distancia <= distancia_maxima:
            return True
    return False

# Determinar distancia máxima (en metros)
distancia_maxima = 500

# Calcular si cada restaurante está cerca de un punto de interés
restaurant_lat_long_puntuacion['cerca_punto_interes'] = restaurant_lat_long_puntuacion.apply(
    lambda row: esta_cerca(row['latitud'], row['longitud'], lugares_interes_lat_long, distancia_maxima), axis=1
)

# Resumen: Proporciones de restaurantes buenos/no buenos cerca o lejos de puntos de interés
resumen = restaurant_lat_long_puntuacion.groupby(['cerca_punto_interes', 'tipo']).size().unstack(fill_value=0)
print("Resumen de restaurantes cerca/lejos de puntos de interés:\n", resumen)

# Porcentajes
resumen_porcentajes = resumen.div(resumen.sum(axis=1), axis=0) * 100
print("\nPorcentajes de restaurantes buenos/no buenos cerca/lejos de puntos de interés:\n", resumen_porcentajes)
```
