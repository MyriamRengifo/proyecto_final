VAMOS A COLOCAR EN UN MAPA LOS LUGARES DE RESTARUANTES Y LOS LUGARES DE INTERES PARA TENER UNA VISTA EN GENERAL DE LA DISTRIBUCION.

```python
restaurant_lat_long = restaurant[['id_nombre','latitud','longitud']][restaurant['puntuacion_yelp'] > 4.5]
lugares_interes_lat_long = lugares_interes[['id_nombre','latitud','longitud']]
```

```python
import folium

# Cargar los datasets
# df_restaurantes = pd.read_csv('restaurant_lat_long.csv')
# df_lugares_interes = pd.read_csv('lugares_interes_lat_long.csv')

# Crear un mapa centrado en una coordenada específica (por ejemplo, la primera de los restaurantes)
mapa = folium.Map(location=[restaurant_lat_long.iloc[0]['latitud'], restaurant_lat_long.iloc[0]['longitud']], zoom_start=12)

# Añadir los restaurantes al mapa
for index, row in restaurant_lat_long.iterrows():
    folium.Marker([row['latitud'], row['longitud']], popup=row['id_nombre'], icon=folium.Icon(color='blue')).add_to(mapa)

# Añadir los lugares de interés al mapa
for index, row in lugares_interes_lat_long.iterrows():
    folium.Marker([row['latitud'], row['longitud']], popup=row['id_nombre'], icon=folium.Icon(color='red')).add_to(mapa)

# Guardar el mapa en un archivo HTML
mapa.save('mapa_restaurantes_lugares_interes.html')
```
