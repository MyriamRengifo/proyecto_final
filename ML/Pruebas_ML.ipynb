VAMOS A UTILIZAR EL METODO LATENT DIRICHLET ALLOCATION (LDA), LA CUAL ES UTILIZADA PARA EL ANALISIS DE TOPICOS. El algoritmo agrupa palabras que tienden a aparecer juntas en los mismos documentos, revelando los "tópicos" ocultos. 

```python
prueba = review_data_test['text']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Ejemplo de preprocesamiento y aplicación de LDA
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(prueba)

LDA_model = LatentDirichletAllocation(n_components=4, random_state=42)
LDA_model.fit(doc_term_matrix)

topics = LDA_model.components_
```

```python
import numpy as np

# Obtener los nombres de las palabras (vocabulario)
vocab = vectorizer.get_feature_names_out()

# Número de palabras más importantes a mostrar por tema
n_top_words = 10

for topic_idx, topic in enumerate(LDA_model.components_):
    print(f"Tema {topic_idx}:")
    # Obtener las palabras más relevantes para el tema
    top_words_idx = topic.argsort()[-n_top_words:]
    top_words = [vocab[i] for i in top_words_idx]
    print(" ".join(top_words))
```
