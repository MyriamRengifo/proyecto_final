```python
import pandas as pd
```

```python
florida_restaurants = pd.read_parquet('../Datasets/florida_restaurants.parquet', engine='pyarrow')
```

```python
florida_restaurants.dtypes
```

Valores duplicados

```python
duplicados_gmap_id = florida_restaurants[florida_restaurants["gmap_id"].duplicated()]
print(len(duplicados_gmap_id))
duplicados_gmap_id
```

```python
# Eliminar duplicados basados en gmap_id
florida_restaurants_df = florida_restaurants.drop_duplicates(subset=['gmap_id']).reset_index(drop=True)

# Verificar la cantidad de filas antes y después
print(f"Antes: {florida_restaurants.shape[0]} filas")
print(f"Después: {florida_restaurants_df.shape[0]} filas")
```

Exportación Valores únicos

```python
florida_restaurants_df.to_parquet('../Datasets/florida_restaurants_df.parquet', engine='pyarrow')
```
